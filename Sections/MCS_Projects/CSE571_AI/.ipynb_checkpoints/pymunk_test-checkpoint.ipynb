```python
import pymunk

# Initialize space
space = pymunk.Space()
space.gravity = (0.0, -900.0)

# Create a body
body = pymunk.Body(1, 1666)
body.position = (50, 100)

# Create a shape and attach to body
shape = pymunk.Circle(body, 10)
space.add(body, shape)

# Simulation loop
for i in range(300):
    space.step(1/50.0)
    print(f"Time step: {i}, Position: {body.position}")
```
